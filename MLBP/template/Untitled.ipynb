```python
import pandas as pd
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
heat_map_test_2 = pd.read_csv("heatmap_2L_test.csv",header=None)
heat_map_test_3 = pd.read_csv("heatmap_3L_test.csv",delimiter='\t',header=None)
heat_map_train_2 = pd.read_csv("heatmap_2L_train.csv",delimiter='\t',header = None)
heat_map_train_3 = pd.read_csv("heatmap_3L_train.csv",delimiter='\t',header=None)
learning_rate_ = [0.01,0.03,0.05,0.1,0.5]
keep_prob_ = [1,0.98,0.95,0.90]
```

```python
print(heat_map_test_2.values)
print(heat_map_test_3.values)
print(heat_map_train_2.values)
print(heat_map_train_3.values)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.heatmap(heat_map_train_2,xticklabels=learning_rate_,yticklabels=keep_prob_)
plt.title("2 Layers, Train set")
plt.xlabel("Learning rate")
plt.ylabel("Dropout")
plt.show()
sns.heatmap(heat_map_test_2,xticklabels=learning_rate_,yticklabels=keep_prob_)
plt.title("2 Layers, Test set")
plt.xlabel("Learning rate")
plt.ylabel("Dropout")
plt.show()
sns.heatmap(heat_map_train_3,xticklabels=learning_rate_,yticklabels=keep_prob_)
plt.title("3 Layers, Train set")
plt.xlabel("Learning rate")
plt.ylabel("Dropout")
plt.show()
sns.heatmap(heat_map_test_3,xticklabels=learning_rate_,yticklabels=keep_prob_)
plt.title("3 Layers, Test set")
plt.xlabel("Learning rate")
plt.ylabel("Dropout")
plt.show()
```
